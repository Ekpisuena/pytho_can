```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df2=pd.read_csv('Total_Revenue.csv') 
print(df2)
```

```python
plt.figure(figsize=(10, 20))
plt.barh(df2['country'], df2['revenue'], color='teal')
plt.title('DVD Rental Revenue per Country', fontsize=14)
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.grid(True)
plt.tight_layout()
```
