```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df1=pd.read_csv('Movie_description.csv')
print('Movie Sentiment Analysis')
def sentiment_detection(description):
  positive = False
  negative = False
  for i in ('Stunning', 'Thrilling','Astounding', 'Beautiful', 'Amazing', 'Brilliant', 'Awe-Inspiring', 'Action-Packed'):
    if i in description:
      positive = True
  for i in ('Sad', 'Scary','Mad', 'Boring', 'Emotional'):
    if i in description:
      negative = True
  if positive == negative == False:
    return "neutral"
  elif positive == True != negative:
    return "positive"
  elif negative == True != positive:
    return 'negative'
  else:
    return "mixed"
df1['Sentiment_Classification'] = df1['description'].apply(lambda x: sentiment_detection(x))
df1
```
