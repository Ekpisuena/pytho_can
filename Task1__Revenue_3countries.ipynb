SQL Query File - Country_Revenue.sql
SQL download file - Revenue.csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df2=pd.read_csv('Revenue.csv') 
df3=df2.loc[(df2['country'] == 'Germany') | (df2['country'] == 'Japan' )| (df2['country'] == 'Philippines' )]
print(df3)
df3.pivot(index="daily", columns="country", values="revenue").plot(figsize=(10,5))
plt.show()
```
